## 사용 컬럼 추출

```python
covid_vaccine = covid_vaccine[["일자", "전체 1차", "전체 2차", "전체 동절기"]]
covid_vaccine
```

## 2021~2022년 데이터 추출

```python
va = covid_vaccine[covid_vaccine["일자"].str.contains("2023")].index
covid_vaccine.drop(va, inplace = True)
covid_vaccine
```

```python
covid_vaccine = covid_vaccine.loc[::-1]
covid_vaccine
```

```python
covid_vaccine.info()
```

## 전체 누적 행 삭제

```python
covid_vaccine = covid_vaccine.drop(674)
covid_vaccine
```

## 누적 백신 접종자 컬럼 추가

```python
covid_vaccine["1차 접종자 누적"] = covid_vaccine["전체 1차"].cumsum()
covid_vaccine["2차 접종자 누적"] = covid_vaccine["전체 2차"].cumsum()
covid_vaccine["동절기 접종자 누적"] = covid_vaccine["전체 동절기"].cumsum()
covid_vaccine
```

## 컬럼명 변경

```python
covid_vaccine.rename(columns = {'전체 1차':'first_shot', '전체 2차':'second_shot', '전체 동절기':'winter_shot',
                                '1차 접종자 누적' : 'accumulate_first_shot',
                                '2차 접종자 누적' : 'accumulate_second_shot',
                                '동절기 접종자 누적' : 'accumulate_winter_shot', '일자' : 'date'},
                     inplace = True)
covid_vaccine
```

## 3차 백신

```python
covid_vaccine_3 = pd.read_csv("./data/covid_vaccine.csv")
covid_vaccine_3 
```

## 데이터 전처리
- 1,2차 코로나 백신 접종자의 경우 정부 데이터 사용
- 3차 백신 접종자 데이터만 추출

## sido 컬럼에서 전국만 추출

```python
covid_vaccine_3 = covid_vaccine_3[covid_vaccine_3["sido"] == "전국"]
covid_vaccine_3
```

## 불필요 컬럼 제거

- 3차 백신 접종자 컬럼만 추출
    - 3차 백신 접종자 누적은 NaN값을 0으로 대체 후 별도의 컬럼으로 추출

```python
covid_vaccine_3 = covid_vaccine_3[["baseDate", "thirdCnt"]]
covid_vaccine_3
```

## 인덱스 수정

```python
covid_vaccine_3 = covid_vaccine_3.reset_index(drop=True)
covid_vaccine_3
```

## 시계열 데이터로 변환

```python
covid_vaccine_3["baseDate"] = pd.to_datetime(covid_vaccine_3["baseDate"])
covid_vaccine_3
```

## 컬럼명 변경

```python
covid_vaccine_3.rename(columns = {'baseDate':'date', 'thirdCnt':'3차 접종자'}, inplace = True)
covid_vaccine_3
```

## 1,2차 데이터 병합

```python
covid_vaccine["date"] = pd.to_datetime(covid_vaccine["date"])
covid_vaccine
```

## 1,2차 백신 접종자에 3차 백신 접종자 병합

```python
total_covid_vaccine = pd.merge(covid_vaccine, covid_vaccine_3, on = 'date', how = 'left')
total_covid_vaccine
```

## NaN값 0으로 대체

```python
total_covid_vaccine = total_covid_vaccine.fillna(0)
total_covid_vaccine = total_covid_vaccine.astype({'3차 접종자' : "int"})
total_covid_vaccine
```

## 3차 백신 접종자 누적 컬럼 추가

```python
total_covid_vaccine["3차 접종자 누적"] = total_covid_vaccine["3차 접종자"].cumsum()
total_covid_vaccine
```

```python
total_covid_vaccine.rename(columns = {"3차 접종자" : "third_shot", "3차 접종자 누적" : "accumulate_third_shot"},
                           inplace = True)
total_covid_vaccine
```

```python
total_covid_vaccine = total_covid_vaccine[["date", "first_shot", "second_shot", "third_shot", "winter_shot",
                                           "accumulate_first_shot", "accumulate_second_shot",
                                           "accumulate_third_shot", "accumulate_winter_shot"]]
total_covid_vaccine
```
